```python
import pandas as pd
from glob import glob
import numpy as np
import os
```

```python
stock_files = sorted(glob('data/stock*.csv'))
stock_files
```

```python
data = pd.concat((pd.read_csv(file).assign(filename = file)
          for  file in stock_files),ignore_index = True)
```
